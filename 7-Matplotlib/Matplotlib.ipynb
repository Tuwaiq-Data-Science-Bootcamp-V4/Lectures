# Overview of the dataset

```python
# import the dataset
df = pd.read_csv('tmdb-movies.csv')
```

```python
# display all columns and how many entries in each column along with their data type
df.info()
```

```python
# delete unnecessary column 
df = df.drop(['cast','homepage', 'director', 'tagline', 'keywords','overview',
              'production_companies','release_date','budget_adj','revenue_adj'], axis = 1)
```

## Line chart 

### How many movies were produced each year?

```python
df.head()
```

```python
film_count_yearly = df.release_year.value_counts(ascending=True)
years = df.groupby('release_year', as_index=False).sum()['release_year']
# as_index
# return object with group labels as the index. Only relevant for DataFrame input. 
# as_index=False is effectively SQL-style grouped output.
```

```python
# Line chart
plt.plot(years, film_count_yearly, c = '0.4')
# c >> grayscale color value in the interval [0, 1] >> '0' as black, '1' as white, '0.8' as light gray

plt.xlabel('Year')
plt.ylabel('Movie Count')
plt.title('Number of Movies Produced every Year')

plt.xlim(1960,2015) # Get or set the x limits of the current axes.
plt.xticks(range(1960,2016,5)) # Get or set the current tick locations and labels of the x-axis.
plt.show()
```

```python
# Line chart
plt.plot(years, film_count_yearly, c = '0.4')
plt.xlabel('Year')
plt.ylabel('Movie Count')
plt.title('Number of Movies Produced every Year')
plt.xlim(1960,2015)
plt.xticks(range(1960,2016,5))
plt.axis('off') # to remove axis
plt.show()
```

```python
# Line chart
plt.plot(range(1960,2015,10), [500,550,430,300,120,100])
plt.plot(years, film_count_yearly, c ='green', linestyle = 'solid', linewidth = 3)
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

```python
# Line chart
plt.plot(range(1960,2011,10), [500,550,430,300,120,100])
plt.plot(years,film_count_yearly, c ='green', linestyle = 'solid', linewidth = 3)
plt.legend(['Cartoon','Movies'], shadow = True, loc = 0) # location (0 = best)
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

```python
# Line chart
plt.plot(range(1960,2011,10), [500,550,430,300,120,100], color = 'blue', linestyle = '--', linewidth = 2)
plt.plot(years,film_count_yearly, c ='green', linestyle = ':', linewidth = 3)
plt.legend(['Cartoon','Movies'], shadow = True, loc = 0) # location (0 = best)
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

```python
# Line chart
plt.plot(range(1960,2011,10), [500,550,430,300,120,100], color = 'blue', linestyle = '--', linewidth = 2)
plt.plot(years,film_count_yearly, c ='green', linestyle = ':', linewidth = 3)
plt.legend(['Cartoon','Movies'], shadow = True, loc = 'center')
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

```python
# Line chart
plt.plot(range(1960,2011,10), [500,550,430,300,120,100], color = 'blue', linestyle = '--', linewidth = 2)
plt.plot(years,film_count_yearly, c ='green', linestyle = ':', linewidth = 3)
plt.legend(['Cartoon','Movies'], shadow = True, loc = (0.2, 0.4))
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

```python
plt.plot(range(1960,2011,10), [500,550,430,300,120,100], marker = 'o', markersize = 15,
         markerfacecolor = 'green', markeredgewidth= 3, markeredgecolor= 'red');
```

```python
# Colors
plt.plot([1,2,0,4,5], 'g')
plt.plot([10,9,8,6,3], c ='lime')
plt.plot([12,13, 10, 20,13], color ='blue')
plt.plot([10,20,30,40,50], '#4B0082')
plt.plot([0, 3, 4, 5 , 9], color ='#A52A2A');
```

## Bar chart 

### Top five movies based on profits? 

```python
top_movies = df.groupby('original_title').sum().sort_values(by='revenue', ascending = False)
top_movies.head()
```

```python
# Bar Chart 
movies = ['Avatar', 'Star Wars', 'Titanic', 'The Avengers', 'Jurassic World']
revenue  = [2781505847, 2068178225, 1845034188, 1568080742, 1513528810]
colors = ['green','blue']

plt.figure(figsize=(10,8))
plt.bar(movies, revenue, color= colors)
plt.title('Top Five Movies',fontsize = 20)
plt.xlabel('Movies',fontsize = 20)
plt.ylabel('Profit',fontsize = 20)
plt.show()
```

### Scatter chart 

```python
plt.scatter(df['vote_average'], df['revenue']) 

plt.title('Top Five Movies',fontsize = 20)
plt.xlabel('Vote Average',fontsize = 20)
plt.ylabel('Profit',fontsize = 20);
```

```python
plt.scatter(df['vote_average'], df['revenue'], alpha =0.3) # alpha between 0 (transparent) and 1 (opaque)
plt.title('Profit vs. Vote Average',fontsize = 20)
plt.xlabel('Vote Average',fontsize = 20)
plt.ylabel('Profit',fontsize = 20);
```

```python
plt.scatter(df['vote_average'], df['revenue'])
plt.scatter(df['vote_count'], df['revenue'])


plt.title('Profit vs. Vote Average & Vote Count', fontsize = 20)
plt.xlabel('Vote Average', fontsize = 20)
plt.ylabel('Profit', fontsize = 20);
```

```python
plt.scatter(df['vote_average'], df['revenue'], marker = '*', s = 100, label = 'Vote average') # s >> size
plt.scatter(df['vote_count'], df['revenue'], marker = '+', s = 100, label ='Vote Count')

plt.legend(loc = 'upper center')
plt.title('Profit vs. Vote Average & Vote Count',fontsize = 20)
plt.xlabel('Vote', fontsize = 20)
plt.ylabel('Profit', fontsize = 20);
```
